```
import pandas as pd
import matplotlib.pyplot as plt
import os


class AmazonAPIClient:
    def __init__(self, api_key):
        self.params = {
            "api_key": api_key,
            "endpoint": "https://api.scraper-service.com/v1",
            "fields": ["price", "rating", "reviews"],
            "data_limit": 50
        }
        print(f"API Client Initialized: {self.params['endpoint']}")


class DataEngine:
    @staticmethod
    def process_csv(file_path):
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File {file_path} not found.")

        df = pd.read_csv(file_path)

        def clean_multi_price(price_val):
            if pd.isna(price_val): return [0.0]
            clean_str = str(price_val).replace('$', '').replace(',', '').strip()
            if '-' in clean_str:
                return [float(p.strip()) for p in clean_str.split('-')]
            try:
                return [float(clean_str)]
            except:
                return [0.0]

        df['price_list'] = df['price'].apply(clean_multi_price)
        df_exploded = df.explode('price_list')

        for col in ['rating', 'reviews']:
            df_exploded[col] = pd.to_numeric(
                df_exploded[col].astype(str).str.replace(r'[^\d.]', '', regex=True),
                errors='coerce'
            ).fillna(0).astype(float)

        df_exploded['price_list'] = df_exploded['price_list'].astype(float)
        return df_exploded[df_exploded['price_list'] > 0]


class Visualizer:
    @staticmethod
    def generate_results(df):
        subset = df.head(15).copy()
        subset['asin'] = subset['asin'].astype(str)

        # Result 1: Multi-Price Line Analysis
        plt.figure(figsize=(10, 6))
        plt.plot(subset['asin'], subset['price_list'], marker='o', color='crimson', label='Price Point')
        plt.title('Result 1: Multi-Price Line Output', fontweight='bold')
        plt.xlabel('Product ID (ASIN)')
        plt.ylabel('Price (USD)')
        plt.xticks(rotation=45)
        plt.grid(True, linestyle='--')
        plt.tight_layout()
        plt.savefig('result_1_price.png')

        # Result 2: Ratings Bar Analysis
        plt.figure(figsize=(10, 6))
        plt.bar(subset['asin'], subset['rating'], color='skyblue', edgecolor='black')
        plt.title('Result 2: Product Ratings Analysis', fontweight='bold')
        plt.ylabel('Rating (Stars)')
        plt.xticks(rotation=45)
        plt.ylim(0, 5.5)
        plt.tight_layout()
        plt.savefig('result_2_ratings.png')

        # Result 3: Review Volume Area Chart
        plt.figure(figsize=(10, 6))
        x_indices = range(len(subset))
        plt.fill_between(x_indices, 0, subset['reviews'], color='green', alpha=0.3)
        plt.plot(x_indices, subset['reviews'], color='darkgreen', marker='s')
        plt.title('Result 3: Customer Reviews Volume', fontweight='bold')
        plt.ylabel('Review Count')
        plt.xticks(x_indices, subset['asin'], rotation=45)
        plt.grid(True, linestyle=':', alpha=0.6)
        plt.tight_layout()
        plt.savefig('result_3_reviews.png')


if __name__ == "__main__":
    API_KEY = "b3a3da6170c6e63c9c72220fc6fcff1ca8a8f68ebe43205d42fed83c243b6380"
    api = AmazonAPIClient(API_KEY)

    try:
        final_data = DataEngine.process_csv('apple_airpods_pro_3.csv')
        Visualizer.generate_results(final_data)
        final_data.to_csv('processed_amazon_data.csv', index=False)
        print("Success: 3 Visualizations generated and data saved.")
    except Exception as e:
        print(f"Error: {e}")
```
